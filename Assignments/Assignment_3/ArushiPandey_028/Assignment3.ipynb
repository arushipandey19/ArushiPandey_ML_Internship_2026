```python
#Question 2
def sum_natural_numbers(n):
  sum_numbers=0
  for i in range(1,n+1):
    sum_numbers+=i
  return sum_numbers
answer=sum_natural_numbers(5)
print(answer)
```

```python
#Question 3
def reverse(n):
  last_digit=0
  rev=0
  while(n>0):
    last_digit=n%10
    rev=rev*10+last_digit
    n=n//10
  return rev
answer=reverse(9235)
print(answer)
```

```python
#Question 4
def count(n):
    cnt=0
    while n>0:
        cnt+=1
        n=n//10
    return cnt
answer = count(9235028710)
print(answer)
```

```python
#Question 5
def palindrome(n):
  last_digit=0
  orig=n
  rev=0
  while(n>0):
    last_digit=n%10
    rev=rev*10+last_digit
    n=n//10
  if rev==orig:
    return True
  else:
    return False
answer=palindrome(92329)
print(answer)
```

```python
#Question 6
def fib(n):
  fib_1=0
  fib_2=1
  for i in range(1,n+1):
    print(fib_1)
    next_term=fib_1+fib_2
    fib_1=fib_2
    fib_2=next_term
fib(6)
```

```python
#Question 7
def calculator():
  n1=int(input("Enter the first number: "))
  n2=int(input("Enter the second number: "))
  ch=int(input("Enter operation (1 for addition, 2 for subtraction, 3 for multiplication, 4 for division): "))
  if ch==1:
    return n1+n2
  elif ch==2:
    return n1-n2
  elif ch==3:
    return n1*n2
  elif ch==4:
    return n1/n2
  else:
    return 0
answer=calculator()
print(answer)
```
